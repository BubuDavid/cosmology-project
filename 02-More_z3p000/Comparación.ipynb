En este notebook, con los resultados obtenidos de las simulaciones, se comparar√° con lo que se puede calcular en nbodykit.

```python
from nbodykit.lab import cosmology
from nbodykit.binned_statistic import BinnedStatistic
from nbodykit.lab import FFTPower

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading our results
Pk = BinnedStatistic.from_json("./results/PowerSpectrum.json")[1:]
```

```python
cosmo = cosmology.Cosmology()
```

```python
# Defining the cosmology and the power spectrum.
c = cosmology.Planck15
Plin = cosmology.LinearPower(c, redshift=3., transfer='CLASS')
```

```python
k = np.logspace(-3, 0.5, 100)
plt.loglog(k, Plin(k), c='k')
plt.loglog(Pk['k'], Pk['power'].real)

plt.xlim(7e-3)
plt.xlabel(r"$k$ $[h \mathrm{Mpc}^{-1}]$")
plt.ylabel(r"$P$ $[h^{-3} \mathrm{Mpc}^{3}]$")
plt.title("Comparaci√≥n entre nbodykit model y la simulaci√≥n")
plt.show()
```
